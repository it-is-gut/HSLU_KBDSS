### Stud Farm Example

```python
bn = gum.loadBN("data/studfarmBN.net")
bn.display()
```

```python
bn.display_tables()
```

```python
ie = gum.LazyPropagation(bn)
```

```python
ie.setEvidence({"Dennis": 2})

for node in bn.nodes():
    ie.setTargets({node})
    ie.makeInference()
    p = ie.posterior(node)
    display(p)
```

### Car Repair Example

```python
bn = gum.loadBN("data/carBN.net")
bn.display()
```

```python
ie = gum.LazyPropagation(bn)
ie.setEvidence({"Car_Starts": 1})

for node in bn.nodes():
    ie.setTargets({node})
    ie.makeInference()
    p = ie.posterior(node)
    display(p)
```

```python
ie = gum.LazyPropagation(bn)
ie.setEvidence({"Car_Starts": 1, "Head_Lights": 0, "Spark_Plugs": 0, "Battery_Age": 0})

for node in bn.nodes():
    ie.setTargets({node})
    ie.makeInference()
    p = ie.posterior(node)
    display(p)
```
